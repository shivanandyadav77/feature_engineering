```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
# to show all the column of the dataframe in the notebook
pd.set_option("display.max_columns",None)
```

```python
#let's load the house price dataset
# and explorer its shape(row and colums)

data=pd.read_csv("./dataset/housing_price/train.csv")
data.shape
```

```python

# lets's visualise the dataset
data.head()
```

```python

# find the variable with missing observations
vars_with_na=[var for var in data.columns if data[var].isnull().mean()>0]
vars_with_na
```

```python
# let's find out whether they are numerical or categorical
data[vars_with_na].dtypes
```

```python
#let's have a look at the values of the variables with missing data

data[vars_with_na].head(10)
```

```python
# let's find out  the percentage of observation missing per variable
# calculate the percentage of missing( as we did in section 3)
# using the isnull() and mean?() methods from pandas

data_na=data[vars_with_na].isnull().mean()

data_na
```

```python
# transform the array into dataframe

data_na=pd.DataFrame(data_na.reset_index())

data_na
```

```python
# adda columns names to the dataframe

data_na.columns=['variable','na_percentage']
data_na
```

```python
#order the dataframe according the percentage of na per variable

data_na.sort_values(by='na_percentage',ascending=False, inplace=True)

#show

data_na
```

```python
# capture variables with no or less than 5%

vars_cca=[var for var in data.columns if data[var].isnull().mean()<0.05]

vars_cca
```
